```python
#q1 = pays les plus pauvres, q4 = pays les plus riches
liste_finale["gdp_quartile"] = pd.qcut(liste_finale["log_gdp_per_capita"],q=4, labels=["q1", "q2", "q3", "q4"])
pays_moyens = liste_finale[liste_finale["gdp_quartile"].isin(["q2", "q3"])]
pays_moyens["life_ladder"].mean()
```

```python
pays_moyens.nunique()
```

```python
liste_finale_gdp = liste_finale[["country_name", "year", "log_gdp_per_capita", "life_ladder"]]
gdp_moyenne = (liste_finale_gdp.groupby("country_name").agg(gdp_mean=("log_gdp_per_capita", "mean"), life_ladder_mean=("life_ladder", "mean"),))
gdp_moyenne
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(gdp_moyenne["gdp_mean"], gdp_moyenne["life_ladder_mean"])
plt.xlabel("PIB moyen (log_gdp_per_capita)")
plt.ylabel("Bonheur moyen (life_ladder)")
plt.title("PIB moyen vs Bonheur moyen (gdp_moyenne)")
plt.show()
```

```python
from scipy.stats import pearsonr

_, p = pearsonr(gdp_moyenne["gdp_mean"], gdp_moyenne["life_ladder_mean"])

print("p-value :", p)
# Plus la P-Value est proche de 0, plus la corrélation est forte
```

```python
np.corrcoef(gdp_moyenne["gdp_mean"], gdp_moyenne["life_ladder_mean"])[0, 1]
# Plus la valeur s'éloigne de 0 et se rapproche de 1 (ou -1), la corrélation est importante
```
